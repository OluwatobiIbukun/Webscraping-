__Job Desc__

```python
outter_div = soup.find('div', {'class': 'css-y4udm8'})
```

```python
len(outter_div.find_all('div'))  # nested_divs
```

```python
outter_div.find_all('div')[1].text
```

```python
# all outer boxes
outter_divs = soup.find_all('div', {'class': 'css-y4udm8'})
```

```python
for out in outter_divs:
    nested_boxes = out.find_all('div')
    second_box = nested_boxes[1]
    
    print(second_box.text)
    print('*' * 50)
```

```python
for out in outter_divs:
    nested_boxes = out.find_all('div')
    second_box = nested_boxes[1]
    
    # levels, yrs
    level = second_box.a
    yrs_exp = second_box.span
    
    print(level.text, ',', yrs_exp.text)
    print('*' * 50)
```
