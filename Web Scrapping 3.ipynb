### Click

1. find_element(By.XPATH, x_path)
2. action ---> ActionChains

```python
from selenium.webdriver.common.by import By  # search
from selenium.webdriver import ActionChains  # actions 
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.amazon.eg/s?rh=n%3A18018102031&fs=true&language=en&ref=lp_18018102031_sar')
```

```python
x_path = '//*[@id="search"]/div[1]/div[1]/div/span[1]/div[1]/div[26]/div/div/span/a[3]'
```

```python
button = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
```

```python
# click the next button
ActionChains(driver).click(button).perform()
```

```python
# 2nd page source code
driver.page_source
```

__click 3rd__

```python
x_path = r'//*[@id="search"]/div[1]/div[1]/div/span[1]/div[1]/div[35]/div/div/span/a[4]'

button = driver.find_element(By.XPATH, x_path)
```

```python
ActionChains(driver=driver).click(button).perform()
```

```python
# 3rd page source code
src = driver.page_source
```

```python
soup = BeautifulSoup(src, 'lxml')
```
